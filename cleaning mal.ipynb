## Standings

## MVP 

```python
url_start = "https://www.basketball-reference.com/awards/awards_{}.html"
```

```python
mvp_players = list(range(1995, 2022))
```

```python
for year in years:
    url = url_start.format(year)
    
    data = requests.get(url)
    
    with open("mvp/{}.html".format(year), "w+") as f:
        f.write(data.text)
```

```python
with open("mvp/1995.html") as f:
    page = f.read()
```

```python
soup_4 = BeautifulSoup(page, 'html.parser')
soup.find('tr', class_="over_header")
```

```python
mvp_players = soup_4.find(id='mvp')
```

```python
mvp_players_final = pd.read_html(str(mvp_players))[0]
mvp_players_final.head(30)
```

```python
dataframes = []
for year in years:
    with open("mvp/{}.html") as f:
        page = f.read()
    soup_4 = BeautifulSoup(page, 'html.parser')
    soup.find('tr', class_="over_header")
    mvp_players = soup_4.find(id='mvp')
    mvp_players_final = pd.read_html(str(mvp_players))[0]
    mvp_players_final['Year'] = year
    
    dataframes.append(mvp_players_final)
```

```python
mvps = pd.concat(dataframes)
```
